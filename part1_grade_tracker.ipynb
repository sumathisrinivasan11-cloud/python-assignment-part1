```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

print("=== TASK 1 OUTPUT ===\n")

for student in raw_students:

    # Clean the name
    name = student["name"].strip().title()

    # Convert roll number
    roll = int(student["roll"])

    # Convert marks_str → list of integers
    marks_text = student["marks_str"].split(",")
    marks = []
    for m in marks_text:
        marks.append(int(m.strip()))

    # Validate name
    words = name.split()
    valid = True
    for w in words:
        if not w.isalpha():
            valid = False

    # Print validation
    if valid:
        print(f"{name}: ✓ Valid name")
    else:
        print(f"{name}: ✗ Invalid name")

    # Profile card
    print("================================")
    print(f"Student : {name}")
    print(f"Roll No : {roll}")
    print(f"Marks   : {marks}")
    print("================================")

    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })

# Special requirement
for s in cleaned_students:
    if s["roll"] == 103:
        print("\n--- Special Output for Roll 103 ---")
        print(f"Name in ALL CAPS : {s['name'].upper()}")
        print(f"Name in lowercase: {s['name'].lower()}")
        print("-----------------------------------")
```

```python
# ------------------------------------------------------------
# TASK 2: MARKS ANALYSIS USING LOOPS & CONDITIONALS
# ------------------------------------------------------------

print("\n=== TASK 2 OUTPUT ===\n")

student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

print("=== SUBJECT-WISE MARKS AND GRADES ===")

# Loop through subjects and marks
for i in range(len(subjects)):
    subject = subjects[i]
    mark = marks[i]

    # Determine grade
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subject}: {mark} → Grade {grade}")

# Calculate total marks
total = 0
for m in marks:
    total = total + m

# Calculate average
average = total / len(marks)

# Find highest and lowest scoring subjects
highest_mark = marks[0]
lowest_mark = marks[0]
highest_subject = subjects[0]
lowest_subject = subjects[0]

for i in range(len(marks)):
    if marks[i] > highest_mark:
        highest_mark = marks[i]
        highest_subject = subjects[i]

    if marks[i] < lowest_mark:
        lowest_mark = marks[i]
        lowest_subject = subjects[i]

print("\n=== SUMMARY ===")
print(f"Total Marks   : {total}")
print(f"Average Marks : {average:.2f}")
print(f"Highest Score : {highest_subject} ({highest_mark})")
print(f"Lowest Score  : {lowest_subject} ({lowest_mark})")

# ------------------------------------------------------------
# WHILE LOOP — ADD NEW SUBJECTS
# ------------------------------------------------------------

print("\n=== ADD NEW SUBJECTS ===")

new_count = 0

while True:
    new_subject = input("Enter subject name (or 'done' to stop): ").strip()

    if new_subject.lower() == "done":
        break

    new_marks_input = input(f"Enter marks for {new_subject}: ").strip()

    # Validate marks
    if not new_marks_input.isdigit():
        print("Invalid marks! Must be a number between 0 and 100.")
        continue

    new_marks = int(new_marks_input)

    if new_marks < 0 or new_marks > 100:
        print("Invalid marks! Must be between 0 and 100.")
        continue

    # Add valid subject + marks
    subjects.append(new_subject)
    marks.append(new_marks)
    new_count = new_count + 1

print("\n=== UPDATED SUMMARY ===")
print(f"New subjects added: {new_count}")

# Recalculate updated average
new_total = 0
for m in marks:
    new_total = new_total + m

new_average = new_total / len(marks)

print(f"Updated Average Marks: {new_average:.2f}")
```

```python
# ------------------------------------------------------------
# TASK 3: CLASS PERFORMANCE SUMMARY
# ------------------------------------------------------------

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("=== CLASS PERFORMANCE REPORT ===\n")
print("Name               | Average | Status")
print("----------------------------------------")

pass_count = 0
fail_count = 0

# To find topper and class average
all_averages = []
topper_name = ""
topper_avg = 0

for student in class_data:
    name = student[0]
    marks = student[1]

    # Calculate average
    total = 0
    for m in marks:
        total = total + m
    avg = total / len(marks)
    avg_rounded = round(avg, 2)

    # Determine pass/fail
    if avg_rounded >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    # Track topper
    if avg_rounded > topper_avg:
        topper_avg = avg_rounded
        topper_name = name

    # Store for class average
    all_averages.append(avg_rounded)

    # Print row
    print(f"{name:<18} | {avg_rounded:>7.2f} | {status}")

# Class average
class_avg = sum(all_averages) / len(all_averages)

print("\n=== SUMMARY ===")
print(f"Students Passed : {pass_count}")
print(f"Students Failed : {fail_count}")
print(f"Class Topper    : {topper_name} ({topper_avg:.2f})")
print(f"Class Average   : {class_avg:.2f}")
```

```python
# ------------------------------------------------------------
# TASK 4: STRING MANIPULATION UTILITY
# ------------------------------------------------------------

essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

print("=== TASK 4 OUTPUT ===\n")

# 1. Strip leading and trailing whitespace
clean_essay = essay.strip()
print("1) Cleaned Essay:")
print(clean_essay)
print()

# 2. Convert to Title Case
title_case_essay = clean_essay.title()
print("2) Essay in Title Case:")
print(title_case_essay)
print()

# 3. Count occurrences of "python" (case-insensitive)
# clean_essay is already lowercase, so .count("python") works
python_count = clean_essay.count("python")
print("3) Count of 'python':", python_count)
print()

# 4. Replace "python" with "Python 🐍"
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("4) Essay with Replacements:")
print(replaced_essay)
print()

# 5. Split into sentences using ". "
sentences = clean_essay.split(". ")
print("5) Sentences List:")
print(sentences)
print()

# 6. Print each sentence on a new line, numbered
print("6) Numbered Sentences:")
for i in range(len(sentences)):
    sentence = sentences[i].strip()

    # Add a period at the end if missing
    if not sentence.endswith("."):
        sentence = sentence + "."

    print(f"{i+1}. {sentence}")
```
